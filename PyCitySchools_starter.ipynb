```python
# Dependencies and Setup
import pandas as pd

import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

school_data_complete.head(30)
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools

total_schools = school_data_complete["school_name"].unique()

number_of_schools = (len(total_schools))

print(f"Number of Schools: {number_of_schools}")

#Calculate the total number of students

total_students = school_data_complete["student_name"]

number_of_students = (len(total_students))

print(f"Number of Students: {number_of_students}")

#Calculate the total budget

budget_per_school = school_data_complete["budget"].unique()

total_budget = budget_per_school.sum()

print(f"Total Budget: {total_budget}")


#Calculate the average math score

average_math = school_data_complete["math_score"].mean()

print(f"Average Math Score: {average_math}")

#Calculate the average reading score

average_reading = school_data_complete["reading_score"].mean()

print(f"Average Reading Score: {average_reading}")


#Create a dataframe to hold the above results

#Optional: give the displayed data cleaner formatting
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)

good_math_score = []

for score in school_data_complete.math_score:
    if score >= 70:
        good_math_score.append(True)
    else:
        good_math_score.append(False)

passing_math = pd.Series(good_math_score)

passing_math


above_70_math_score = school_data_complete[passing_math]

len(above_70_math_score)

percent_math_pass = len(above_70_math_score) *100 / len(total_students)


TF_Math = pd.DataFrame(good_math_score)

print(f"Percent of Students Passing Math: {percent_math_pass}")
```

```python
#Calculate the percentage of students with a passing reading score (70 or greater)

good_reading_score = []

for score_r in school_data_complete.reading_score:
    if score_r >= 70:
        good_reading_score.append(True)
    else:
        good_reading_score.append(False)

passing_reading = pd.Series(good_reading_score)


above_70_reading_score = school_data_complete[passing_reading]

len(above_70_reading_score)

percent_reading_pass = len(above_70_reading_score) *100 / len(total_students)


percent_passing_reading = len(good_reading_score)

#convert to dataframe
TF_Reading = pd.DataFrame(good_reading_score)

print(f"Percent of Students Passing Reading: {percent_reading_pass}")
```

```python
#Create new dataframe with Math and Reading T/F

math_tf_df = pd.DataFrame({"Math_TF": good_math_score})

reading_tf_df = pd.DataFrame({"Reading_TF": good_reading_score})

school_data_complete["Reading_TF"] = reading_tf_df
school_data_complete["Math_TF"] = math_tf_df
school_data_complete

```

```python
#Calculate the percentage of students who passed math and reading (% Overall Passing)
passing_overall = []

school_data_complete["Passing_Overall"] = np.logical_and(school_data_complete["Reading_TF"], school_data_complete["Math_TF"] == True)


passing_overall = school_data_complete["Passing_Overall"]

number_passing_overall = passing_overall.sum()



percent_pass_overall = number_passing_overall* 100 / len(total_students)

percent_pass_overall

print(f"Percent of Students Passing Overall: {percent_pass_overall}")


school_data_complete
```

```python
#Create a dataframe to hold the above results

school_data_summary_df = pd.DataFrame({ "Total Schools": number_of_schools,
                                       "Total Students": number_of_students,
                                       "Total Budget": total_budget,
                                       "Average Math Score": average_math,
                                       "Average Reading Score": average_reading,
                                       "Passing Math %": percent_math_pass,
                                       "Passing Reading %": percent_reading_pass,
                                       "Passing Overall %": [percent_pass_overall]
                                      })

school_data_summary_df
```

## Math Scores by Grade

```python
# Loc and groupy for 9th grade math
math_ninth = school_data_complete.loc[school_data_complete["grade"] == "9th"]

math_ninth_grouped = math_ninth.groupby("school_name")["math_score"].mean().to_frame(name = "9th").reset_index()
math_ninth_grouped

# Loc and groupy for 10th grade math

math_tenth = school_data_complete.loc[school_data_complete["grade"] == "10th"]

math_tenth_grouped = math_tenth.groupby("school_name")["math_score"].mean().to_frame(name = "10th").reset_index()
math_tenth_grouped

# Loc and groupy for 11th grade math

math_eleventh = school_data_complete.loc[school_data_complete["grade"] == "11th"]

math_eleventh_grouped = math_eleventh.groupby("school_name")["math_score"].mean().to_frame(name = "11th").reset_index()
math_eleventh_grouped

# Loc and groupy for 12th grade math

math_twelfth = school_data_complete.loc[school_data_complete["grade"] == "12th"]

math_twelfth_grouped = math_twelfth.groupby("school_name")["math_score"].mean().to_frame(name = "12th").reset_index()
math_twelfth_grouped

#Merge Dataframes

avg_math_by_grade910 = pd.merge(math_ninth_grouped, math_tenth_grouped, on="school_name")
avg_math_by_grade911 = pd.merge(avg_math_by_grade910, math_eleventh_grouped, on="school_name")
avg_math_by_grade912 = pd.merge(avg_math_by_grade911, math_twelfth_grouped, on="school_name")
avg_math_by_grade912

#Set Index to school name

school_avgmath_grade = avg_math_by_grade912.set_index("school_name")
school_avgmath_grade
```

```python
#Create bins for school spending

bins = [0, 585, 630, 645, 680]

group_names = ["<$585", "$585-630", "$630-645", "$645-680"]

school_budget_bins_df = school_data_final[["Per Student Budget","Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing"]]

school_budget_bins_df

school_budget_bins_df["Spending Ranges (Per Student)"] = pd.cut(school_budget_bins_df["Per Student Budget"], bins, labels=group_names, include_lowest=True)
school_budget_bins_df
```
